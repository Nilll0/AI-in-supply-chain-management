```python
# load packages
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
MSFT = data.DataReader('MSFT', start='2010-01-01', 
                       end='2019-12-31', data_source='yahoo')
MSFT.head()
```

Note that theree are a number of availabe sources for financial data, such as Yahoo finance and Google Finance.

```python
# we look at MSFT closing price
MSFT_close = MSFT['Close']
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(MSFT_close, label='MSFT', color='green', linewidth=2)
ax.set_xlabel('Date', size=16)
ax.set_ylabel('Closing price ($)', size=16)
ax.legend();
```

```python
MSFT_MA50 = MSFT['Close'].rolling(50).mean()
MSFT_MA100 = MSFT['Close'].rolling(100).mean()
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(MSFT_close, label='MSFT', color='g', linewidth=2)
ax.plot(MSFT_MA50, label='MA50', color='b')
ax.plot(MSFT_MA100, label='MA100', color='orange')

ax.set_xlabel('Date', size=16)
ax.set_ylabel('Closing price ($)', size=16)
ax.legend();
```

```python
start_date = '2019-07-01'
end_date = '2019-10-1'
```

```python
MSFT_2019 = MSFT_close[start_date:end_date]
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(MSFT_2019, label='MSFT', color='g', linewidth=4)

ax.set_xlabel('Date', size=16)
ax.set_ylim([125,150])
ax.set_ylabel('Closing price ($)', size=16)
ax.legend();
```
